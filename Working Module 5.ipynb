```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("datasets/diamonds.csv") #1st column needs encoding to read pound/hashtag(#) symbol
df.head()
```

```python
df.rename(columns={'Unnamed: 0':'Number'}, inplace =True)
```

```python
df.head()
```

```python
df.set_index('Number',inplace=True)
df.head()
```

```python
sns.lmplot(x='carat', y='depth', data=df)
```

```python
pkmn_type_colors=['#98d8d8','#7038f8']
```

```python
plt.subplots(figsize=(13,7))
sns.violinplot(x='color', y='depth', data=df,palette=pkmn_type_colors)
```

```python
plt.subplots(figsize=(10,6))
df_premium = df.loc[df['cut'] == 'Premium']
df2 = df_premium.head(100)
sns.swarmplot(x='color',y='depth',data=df2,palette=pkmn_type_colors)
# sns.swarmplot?
# df.head(15)
```

```python
plt.figure(figsize=(13,7))
sns.violinplot(x='color',
               y='depth',
               data=df2,
               inner=None, #removes bars inside violin plot
               palette=pkmn_type_colors)


```
